```python
# 1. Importing Necessary Libraries
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# 2. Loading the Dataset
df_gender_submission = pd.read_csv('/content/gender_submission.csv')
```

```python
df_gender_submission
```

```python
#3. Summarizing the dataset
print(df_gender_submission.head())
print(df_gender_submission.describe())
```

```python
# 4. Pre-Processing the Dataset
# Checking the missing values
print(df_gender_submission.isnull().sum())
```

Dataset is already clean, no NaN values or missing data to handle.

```python
# 5. Segregating the Dataset into Input (X) and Output (y)
X = df_gender_submission.drop('Survived', axis=1)
y = df_gender_submission['Survived']
```

```python
X
```

```python
y
```

```python
# 6. Performing Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Scaling the data ensures that numerical features are on a similar scale so the SVM algorithm performs better.

```python
X_scaled
```

```python
# 7. Splitting the Dataset into Training and Testing Data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

SPlitting the data into training 80% and testing 20%.

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 8. Loading the Model (SVM in this case)
model = SVC(kernel='sigmoid', random_state=42)
```

I have used Sigmoid kernel function because i dint get y pred as classifying 0 and 1. So , i did trail and error and found sigmoid function gives be better y pred values.

```python
# 9. Training the Model
model.fit(X_train, y_train)
```

```python
# 10. Predicting the Result using the Trained Model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# 11. Calculating the Accuracy of the Trained Model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the SVM model: {accuracy * 100:.2f}%")
```

The accuracy of this model is 55.95%.  



```python
# 12. Predicting the Output of Single Sample using the Trained Model
# Let's use the first sample from the test data as an example
sample = X_test[0].reshape(1, -1)
predicted_class = model.predict(sample)
print(f"Predicted class for the sample: {predicted_class[0]}")
print(f"Actual class for the sample: {y_test.values[0]}")
```

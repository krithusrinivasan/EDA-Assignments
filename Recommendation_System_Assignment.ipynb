```python
import numpy as np
import pandas as pd
```

```python
movies_data = pd.read_csv("/content/movies.csv")
movies_data
```

```python
#number of unique users
# rename 'rank' column name, 'rank' is function in Numpy so need to rename it.
movies_data_copy = movies_data.rename(columns={'rank':'user_rating'})
len(movies_data_copy.user_rating.unique())
```

```python
movies_data_copy.info()
```

```python
len(movies_data_copy.title.unique())
# length of rows is 250
```

```python
user_movies_data = movies_data_copy.pivot(index='user_rating',
                                 columns='title',
                                 values='imbd_rating')
user_movies_data
#Selecting particular columns from the dataset
```

```python
#Impute those NaNs with 0 values
user_movies_data.fillna(0, inplace=True)
user_movies_data
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_movies_data.values, metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids
user_sim_df.index = user_movies_data.index
user_sim_df.columns = user_movies_data.index
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```
